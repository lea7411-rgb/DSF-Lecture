**4.    What are the different types of machine learning algorithms?**

Types of ML:

1.   **Supervised Learning:** learning from labeled data, input -> correct output,
goal is to predict outcomes
*   Common tasks: Classification, Regression

*   Algorithms: Linear and Logistic Regression, Decision Trees, Random Forest, Support Vector Machines SVM, Neural Networks

2.   **Unsupervised Learning:** learning with unlabed data, finds patterns or structure

*   Common tasks: Clustering, Dimensionality reduction

*   K-Means Clustering, Hierarchical Clustering, DBSCAN, Principal Component Analysis PCA

3.  **Reinforcement Learning:** learning by interaction with environment, rewards and penalties, Learining through trial and error

*   Common tasks: Game playing, Robotics, Autonomous driving

*   Algorithms: Q-Learning, Deep Q Networks DQN, Policy Gradient Methods

4.    **Semi-Supervised Learning:** small amount of labeled data + lots of unlabeled data

*   Examples: Predicting missing words in a sentence or next frame in a video e.g. Generative Pre-trained Transformer GPT

5.    **Ensemble Learning:** multiple models, many weak learners -> one strong learner, using several models in combination to improve prediction accuracy

*   Bagging: Random Forest

*   Boosting: AdaBoost, Gradient Boosting

*   Stacking

**6.    How do you evaluate the performance of a machine learning model?**

1.  Train/Test split: ensures model isn't just memodrizing/overfitting

*   Training set -> used to learn patterns

*   Validation set -> used to tune hyperparameters

*   Test set -> used for final evaluation, never touched during training

2.  Metrics for Classification: predicting categories

*   Acuracy: overall correctness = TP + TN / Total, balanced datasets

*   Precision: How many predicted positives are correct = TP / TP + FP, avoids false positives

*   Recall/Sensitivity: How many actual positives are detected =. TP / TP + FN, avoids false negatives

*   F1 Score: Balance between precision & recall

*   ROC-AUC: Mesures model's ability to distinguish classes, Higher = better separation

3.  Metrics for Regression: prediction of numbers

*   Mean Absolute Error MAE: Average absolute difference

*   Mean Squared Error MSE: Penalizes large error more

*   Root Mean Squared Error RMSE: Interppretable, same units as target

*   R-squared: Proportion of variance explained

4.    Cross-Validation: Reduces randomness and gives more robust performance

*   Split data into k folds

*   Train/Test multiple times

*   Average results

5.  Overfitting vs Underfitting: compare training vs validatio performance

*   Overfitting: great on training, poor on test
