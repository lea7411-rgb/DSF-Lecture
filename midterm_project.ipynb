**4.    What are the different types of machine learning algorithms?**

5.    **Ensemble Learning:** multiple models, many weak learners -> one strong learner, using several models in combination to improve prediction accuracy

**6.    How do you evaluate the performance of a machine learning model?**

1.  Train/Test split: ensures model isn't just memodrizing/overfitting

*   Training set -> used to learn patterns

*   Validation set -> used to tune hyperparameters

*   Test set -> used for final evaluation, never touched during training

2.  Metrics for Classification, Confusio Matrix Analysis: Identify where model fails

*   Acuracy: overall correctness = (TP + TN) / Total, balanced datasets

*   Precision: How many predicted positives are correct = TP / (TP + FP), avoids false positives

*   Recall/Sensitivity: How many actual positives are detected =. TP / (TP + FN), avoids false negatives

*   F1 Score: Balance between precision & recall

*   ROC-AUC: Mesures model's ability to distinguish classes, Higher = better separation

3.  Metrics for Regression:

*   Mean Absolute Error MAE: Average absolute difference

*   Mean Squared Error MSE: Penalizes large error more

*   Root Mean Squared Error RMSE: Interppretable, same units as target

*   R-squared: Proportion of variance explained

4.    Cross-Validation: Reduces randomness and gives more robust performance

*   Split data into k folds

*   Train/Test multiple times

*   Average results

5.  Overfitting vs Underfitting: compare training vs validatio performance

*   Overfitting: great on training, poor on test

*   Underfitting: poor everywhere

6.  Learning Curves: Helps diagnos vias/variance issues

*   Show performace vs training size

**7.    How do you prepare data for use in a machine learning model?**

1.  Understand the problem, as it determines how data needs to be prepared

2.  Making Sure data is Relevant, sufficient in size and representative of real-world scenarios

3.  Cleaning the data as raw data can be messy

*   Handling missing values: remove rows or fill (mean, median, interpolation)

*   Removing duplicates

*   Fixing errors: Wrong formats, typos, inconsitent labels

*   Outlier detection: Remove or cap extreme values if needed

*   Remove unnecessary data

4.  Explore the data: Use statistics and visualitations to understand patterns and make better preprocessing decisions

*   Distributions (histograms)

*   Relationships (correlation, scatter plots)

*   class imbalance

5. Feature Engineering: Transform raw data into useful inputs (features):

*   Encoding categrocial variables: One-hot encoding, label encoding

*   Scaling numerical data: Standardization, Normalization

*   Creating new features: Ratios, interactions, domain-specific transformations

*   Date/time features: Extract day, month, weekday etc

6.  Split the Data: Avoid data leakage (never let test data influence training)

*   Training set (70-80%): used to train the model

*   Validation set (optional): tuning

*   Test set (20-30%): final evaluation of the model

7. Handle Imbalanced Data (if needed, if one class dominates)

*   Oversampling (e.g. SMOTE)

*   Undersampling

*   Class weighting

8.  Convert to Model-Ready Format: Depending on Framework (e.g. TensorFlow or PyTorch):

*   Convert to arrays/tensors

*   Ensure consisten shapes

*   Batch the data

9.  Build a Data Pipeline: Automate processing

*   tools like scikit-learn pipelines

*   ensure the same transformations are applied to training and future data

10. Validate the Data before training

*   Check distribution again

*  verifiy no leakage

*   Confirm labels are correct

**8.    What are some common challenges in machine learning, and how can they be addressed?**

1. Poor Data Quality: Missing values, Noisy or inconsistent data, Incorrect labels

* Clean and preprocess data (imputation, normalization)

* Use robust data validation pipelines
* Apply outlier detection methods
* Invest in better labeling (human-in-the-loop if needed)

2.  Overfitting vs. Underfitting: Overfitted Model memorizes training data but fails on new data while Underfitted Model is too simple to capture patterns

* Use train/validation/test splits

* Apply regularization (L1/L2)
* Use cross-validation
* Adjust model complexity (e.g., depth of trees, layers in neural nets)

3.  Data Imbalance: When one class dominates (e.g., fraud detection), models become biased toward it.

* Resampling techniques (oversampling/undersampling)
* Use synthetic data generation (e.g., SMOTE)
* Apply class-weighted loss functions
* Evaluate with appropriate metrics (F1-score, ROC-AUC—not just accuracy)

4. Feature Engineering & Selection: Choosing the right input features is often harder than building the model itself.

* Domain knowledge is critical
* Use automated feature selection (e.g., recursive feature elimination)
* Apply dimensionality reduction (e.g., PCA)
* Use models that handle features well (e.g., tree-based models)

5.  Model Interpretability: Complex models (like deep neural networks) act as “black boxes.”

* Use interpretable models when possible (linear models, decision trees)
* Apply explainability tools: SHAP values, LIME
* Visualize feature importance

6.  Scalability & Performance: Models may work in notebooks but fail at scale (large datasets or real-time use).

* Use distributed computing (e.g., Spark)
* Optimize models (quantization, pruning)
* Deploy with efficient serving systems (e.g., TensorFlow Serving, ONNX)

7. Data Drift & Concept Drift: Real-world data changes over time, making models degrade.

* Monitor model performance continuously
* Set up retraining pipelines
* Use online learning or adaptive models
* Track data distribution changes

8.  Evaluation & Metrics Selection: Choosing the wrong metric leads to misleading results.

* Align metrics with business goals
* Use multiple metrics (precision, recall, F1, AUC)
* Perform error analysis—not just aggregate scores

9.  Bias & Fairness: Models can inherit biases from data, leading to unfair outcomes.

* Audit datasets for bias
* Use fairness-aware algorithms
* Evaluate across subgroups
* Ensure transparency and accountability

10. Deployment & Maintenance (MLOps): Moving from prototype to production is often the hardest step.

* Use MLOps practices: CI/CD for models, Versioning (data + models), Monitoring pipelines
* Automate workflows
* Use tools like MLflow, Kubeflow

**Creating environment**

Ubelix (Linux HPC cluster) has a specific Python version, package versions and system libraries. To avoid or atleast minimize future compatibility issues the local environment is set up to match the one on Ubelix as close as possible.
This is especially important for packages such as Tensorflow but also for numpy, scikit-kit learn and rdkit.  

````python
# creates new environment with Ubelix compatible python version
$ conda create -n dsf_mtp python=3.10
$ conda activate dsf_mtp

# conda-forge automatically resolves pagacke imcompatibility
$ conda install -c conda-forge pandas numpy scikit-learn seaborn rdkit

# application of pip to avoid broken conda TF builds on macOS
# newest tensorflow version which is compatible with Apple Silicon and Ubelix (2.16.1) -> resolves Binary incompatibility
$ pip install tensorflow-macos==2.16.2

# GPU acceleration
$ pip install tensorflow-metal

# adds environment to jupyter
$ pip install ipykernel
$ python -m ipykernel install --user --name dsf_mtp --display-name "Python (dsf_mtp)"

# creates a reproducible environment file
$ conda env export > environment.yml
# or
$ pip freeze > requirements.txt
````
