3.    **What are the key concepts and techniques in machine learning?**

Types of ML:

*   Supervised Learning: learning from labeled data, input -> correct output,
goal is to predict outcomes

*   Unsupervised Learning: learning with unlabed data, finds patterns or structure

*   Reinforcement Learning: learning by interaction with environment, rewards and penalties

Key concepts:

*   Features and labes

*   training vs testing

*   overfitting vs underfitting

*   Bias-Variance Tradeoff: balance between simplicity and flexibility

*   Loss Functions: MSE

Core Techniques:

*   Feature Engineering: Selecting, transforming, or creating input features

*   Model Training/Optimization: Adjust model parameters to minimize loss

*   Regularizarion: Prevents overfitting, Techniques: L1, L2, Dropout in neural networks

*   Cross-Validation: splits data into multiple folds, ensures model is robust
