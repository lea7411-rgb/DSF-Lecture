**4.    What are the key concepts and techniques in machine learning?**

**Data Representation and Features**

Effective data representation uses various formats, such as written, classified, or numerical, to convey information. Features, specific characteristics or attributes, which are used by the model to predict targets/labels, are extracted from this data. Feature engineering involves selecting, modifying, or creating features to improve model performance and identify essential trends.

**Training, Testing, and Validation**

Datasets are typically divided into three subsets to assess a model's performance:

* Training Set: Used to train the model by learning patterns from the data.
During this phase, model parameters are optimized to minimize a chosen loss function, such as Mean Squared Error (MSE) in regression tasks.

* Validation Set: Used to tune hyperparameters, select the best model configuration during training, and monitor issues such as overfitting (model too complex, memorizing data/high variance) and underfitting (model too simple, failing to capture patterns/high bias). The Bias–Variance Tradeoff is a key concept in model development.

* Test Set: Used to evaluate the final model's performance on unseen data after training is complete, providing an unbiased estimate of generalization.

Performance is measured using metrics such as Accuracy, Precision, Recall, and F1 Score.

**Model Selection and Regularization**

* Model Selection: The process of choosing the best architecture for a specific problem, influenced by data type, computational resources, and interpretability needs.
* Regularization: Techniques used to resolve the bias-variance trade-off and prevent overfitting. Methods like L1 and L2 regularization add penalty terms to the loss function to encourage simpler models.


**An additional technique is Cross-Validation**

A method that splits data into multiple folds to train and validate the model repeatedly, ensuring robustness and reducing dependence on a single validation split.

**5.    What are the different types of machine learning algorithms?**

*   **Supervised Learning**   
Models are trained on labeled data where the output is known, allowing the computer to categorize or predict new events.

multiple models, many weak learners -> one strong learner, using several models in combination to improve prediction accuracy

A subset of machine learning using multi-layered **Artificial Neural Networks** to learn complex representations. It is widely used in computer vision, natural language processing (NLP), and speech recognition.

**7.    How do you evaluate the performance of a machine learning model?**

*  Train/Test split: ensures model isn't just memodrizing/overfitting

    *   Training set -> used to learn patterns
    *   Validation set -> used to tune hyperparameters
    *   Test set -> used for final evaluation, never touched during training

*  Metrics for Classification, Confusio Matrix Analysis: Identify where model fails

    *   Acuracy: overall correctness = (TP + TN) / Total, balanced datasets
    *   Precision: How many predicted positives are correct = TP / (TP + FP), avoids false positives
    *   Recall/Sensitivity: How many actual positives are detected =. TP / (TP + FN), avoids false negatives
    *   F1 Score: Balance between precision & recall
    *   ROC-AUC: Mesures model's ability to distinguish classes, Higher = better separation

*  Metrics for Regression:

    *   Mean Absolute Error (MAE): Average absolute difference
    *   Mean Squared Error (MSE): Penalizes large error more
    *   Root Mean Squared Error (RMSE): Interppretable, same units as target
    *   R-squared: Proportion of variance explained
    *   Mean Absolute Percentage Error (MAPE): Relative absolute difference

*  Overfitting vs Underfitting: compare training vs validation performance

    *   Overfitting: great on training, poor on test
    *   Underfitting: poor everywhere

*  Learning Curves: Helps diagnos bias/variance issues

    *   Show performace vs training size

**9.    What are some common challenges in machine learning, and how can they be addressed?**

* Poor Data Quality: Missing values, Noisy or inconsistent data, Incorrect labels

    * Apply regularization (L1/L2)
    * Use cross-validation
    * Adjust model complexity (e.g., depth of trees, layers in neural nets)

*  Data Imbalance: When one class dominates (e.g., fraud detection), models become biased towards it.

    * Resampling techniques (oversampling/undersampling)

    * Apply class-weighted loss functions
    * Evaluate with appropriate metrics (F1-score, ROC-AUC—not just accuracy)

    * Domain knowledge is critical

    * Apply dimensionality reduction (e.g., PCA)
    * Use models that handle features well (e.g., tree-based models)

    * Use interpretable models when possible (linear models, decision trees)
    * Apply explainability tools: SHAP values, LIME
    * Visualize feature importance

*  Scalability & Performance: Models may work in notebooks but fail at scale (large datasets or real-time use).

    * Use distributed computing (e.g., Spark)
    * Optimize models (quantization, pruning)
    * Deploy with efficient serving systems (e.g., TensorFlow Serving, ONNX)

*   Computatuional Cost: High resource requirements

    * Specialized hardware (GPUs)
    * High-performance clusters (e.g. Ubelix)


*  Bias & Fairness: Models can inherit biases from data, leading to unfair outcomes.

    * Audit datasets for bias

    * Ensure transparency and accountability

**6. Create function that converts the chemical structure (as SMILES string) into a numerical fingerprint vector.**  
Function was made more robust by ading lines 20-21 and 26-27. Smiles column can contain missing values (NaN -> float), this will cause an error as RDKit cannot parse a float and expects str. Further smiles string can simply be invalid.

````python
def smiles_to_fp(smiles, method="maccs", n_bits=2048):
    """
    Encode a molecule from a SMILES string into a fingerprint.

    Parameters
    ----------
    smiles : str
        The SMILES string defining the molecule.

    method : str
        The type of fingerprint to use. Default is MACCS keys.

    n_bits : int
        The length of the fingerprint.

    Returns
    -------
    array
        The fingerprint array.
    """
    # handles bad data types (float(NaN), NoneType, int)
    if not isinstance(smiles, str):
        return None
    # Convert smiles to RDKit mol object
    mol = Chem.MolFromSmiles(smiles)

    # Handles invalid smiles strings
    if mol is None:
        return None

    if method == "maccs":
        return np.array(MACCSkeys.GenMACCSKeys(mol))
    if method == "morgan2":
        fpg = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=n_bits)
        return np.array(fpg.GetCountFingerprint(mol))
    if method == "morgan3":
        fpg = rdFingerprintGenerator.GetMorganGenerator(radius=3, fpSize=n_bits)
        return np.array(fpg.GetCountFingerprint(mol))
    else:
        print(f"Warning: Wrong method specified: {method}." " Default will be used instead.")
        return np.array(MACCSkeys.GenMACCSKeys(mol))
````

**10. Defining a keras model (copied verbatim from tutorial).**

````python
def neural_network_model(hidden1, hidden2):
    """
    Creating a neural network from two hidden layers
    using ReLU as activation function in the two hidden layers
    and a linear activation in the output layer.

    Parameters
    ----------
    hidden1 : int
        Number of neurons in first hidden layer.

    hidden2: int
        Number of neurons in second hidden layer.

    Returns
    -------
    model
        Fully connected neural network model with two hidden layers.
    """

    model = Sequential()
    # First hidden layer
    model.add(Dense(hidden1, activation="relu", name="layer1"))
    # Second hidden layer
    model.add(Dense(hidden2, activation="relu", name="layer2"))
    # Output layer
    model.add(Dense(1, activation="linear", name="layer3"))

    # Compile model
    model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse", "mae"])
    return model
````

**11. Trying different mini-batch sizes and plotting the respective losses (copied verbatim from tutorial, except for y-axis limits and matrices).**

````python
# Neural network parameters
batch_sizes = [16, 32, 64]
nb_epoch = 50
layer1_size = 64
layer2_size = 32

# Plot
fig = plt.figure(figsize=(12, 6))
sns.set(color_codes=True)
for index, batch in enumerate(batch_sizes):
    fig.add_subplot(1, len(batch_sizes), index + 1)
    model = neural_network_model(layer1_size, layer2_size)

    # Fit model on x_train, y_train data
    history = model.fit(
        x_train,
        y_train,
        batch_size=batch,
        validation_data=(x_test, y_test),
        verbose=0,
        epochs=nb_epoch,
    )
    plt.plot(history.history["loss"], label="train")
    plt.plot(history.history["val_loss"], label="test")
    plt.legend(["train", "test"], loc="upper right")
    plt.ylabel("loss")
    plt.xlabel("epoch")
    plt.ylim((0, 2))
    plt.title(
        f"test loss = {history.history['val_loss'][nb_epoch-1]:.2f}, " f"batch size = {batch}"
    )
plt.show()
````

**6. Fingerprint function.**

````python
def smiles_to_fp(smiles, method="maccs", n_bits=2048):
    """
    Encode a molecule from a SMILES string into a fingerprint.

    Parameters
    ----------
    smiles : str
        The SMILES string defining the molecule.

    method : str
        The type of fingerprint to use. Default is MACCS keys.

    n_bits : int
        The length of the fingerprint.

    Returns
    -------
    array
        The fingerprint array.
    """
    # handles bad data types (float(NaN), NoneType, int)
    if not isinstance(smiles, str):
        return None
    # Convert smiles to RDKit mol object
    mol = Chem.MolFromSmiles(smiles)

    # molecule validation to avoid chrashing incase mol = None
    if mol is None:
        return None  # or np.zeros(n_bites)

    if method == "maccs":
        return np.array(MACCSkeys.GenMACCSKeys(mol))
    elif method == "morgan2":
        fpg = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=n_bits)
        return np.array(fpg.GetCountFingerprint(mol))
    elif method == "morgan3":
        fpg = rdFingerprintGenerator.GetMorganGenerator(radius=3, fpSize=n_bits)
        return np.array(fpg.GetCountFingerprint(mol))
    else:
        print(f"Warning: Wrong method specified: {method}." " Default will be used instead.")
        return np.array(MACCSkeys.GenMACCSKeys(mol))
````

**9. Neural network model**


````python
def neural_network_model(hidden1, hidden2):
    """
    Creating a neural network from two hidden layers
    using ReLU as activation function in the two hidden layers
    and a linear activation in the output layer.

    Parameters
    ----------
    hidden1 : int
        Number of neurons in first hidden layer.

    hidden2: int
        Number of neurons in second hidden layer.

    Returns
    -------
    model
        Fully connected neural network model with two hidden layers.
    """

    model = Sequential()
    # First hidden layer
    model.add(Dense(hidden1, activation="relu", name="layer1"))
    # Second hidden layer
    model.add(Dense(hidden2, activation="relu", name="layer2"))
    # Output layer
    model.add(Dense(1, activation="linear", name="layer3"))

    # Compile model
    model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse", "mae"])
    return model
````
